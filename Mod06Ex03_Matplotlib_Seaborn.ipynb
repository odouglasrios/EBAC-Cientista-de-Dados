#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
asfixia_grave = (df['APGAR5'] >= 0) & (df['APGAR5'] <= 3)
asfixia_moderada = (df['APGAR5'] >= 4) & (df['APGAR5'] <= 5)
asfixia_leve = (df['APGAR5'] >= 6) & (df['APGAR5'] <=7)
normal = df['APGAR5'] >= 8

df.loc[asfixia_grave, 'categoria'] = 'asfixia_grave'
df.loc[asfixia_moderada, 'categoria'] = 'asfixia_moderada'
df.loc[asfixia_leve, 'categoria'] = 'asfixia_leve'
df.loc[normal, 'categoria'] = 'normal'
```

```python
df['categoria'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(x = 'APGAR5',
           y = 'GESTACAO',
           hue = 'SEXO',
           kind = 'bar',
           data = df,
           ci = False)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_percentual_linha = freq.div(freq.sum(axis = 1), axis = 0)
freq_percentual_linha
```
