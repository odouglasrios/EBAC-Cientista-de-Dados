````python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
# Carregando a base de dados
df = pd.read_csv('previsao_de_renda.csv')
````

````python
# Removendo a coluna 'Unnamed: 0'
df.drop('Unnamed: 0', axis = 'columns', inplace = True)
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Posse de imóvel
ax = sns.countplot(x = 'data_ref', hue = 'posse_de_imovel', data = df);

tick_labs = df['data_ref'].unique()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

title = ax.set_title('POSSE DE IMÓVEL')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
